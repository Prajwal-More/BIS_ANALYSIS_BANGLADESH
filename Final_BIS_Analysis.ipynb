### **Importing Data**

```python
titles = list()
list_of_tables = list()
```

### **ANALYSIS**

Presence of Building

```python
POB = data['Is there a building?'].value_counts()

presence_of_building = pd.DataFrame(POB)

presence_of_building.loc['Total'] = presence_of_building.sum(axis=0)

titles.append("Presence of a Building")
list_of_tables.append(presence_of_building)
```

```python
presence_of_building
```

Type of Building

```python
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Under construction ', 'Under construction' )
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Under Construction', 'Under construction' )
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Under Construcion', 'Under construction' )
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Understand construction ', 'Under construction' )

data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Bamboo & tin', 'Tin & bamboo' )
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Tin and bamboo ', 'Tin & bamboo' )

data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Tin & mud', 'Tin and mud')
data['What is the type of building? (Other (please specify)) - specify'] = data['What is the type of building? (Other (please specify)) - specify'].replace('Toylet', 'Toilet' )
```

```python
TOB = data['What is the type of building?'].value_counts()

Type_of_Building = pd.DataFrame(TOB)
Type_of_Building = Type_of_Building.transpose()

Type_of_Building['Total'] = Type_of_Building.sum(axis=1)

titles.append("Type of a Building")
list_of_tables.append(Type_of_Building)
```

```python
Type_of_Building
```

```python
data['What is the type of building?'].unique()
```

```python
TOB_Others = data['What is the type of building? (Other (please specify)) - specify'].value_counts()

Type_of_Building_others = pd.DataFrame(TOB_Others)
Type_of_Building_others = Type_of_Building_others.transpose()

Type_of_Building_others['Total'] = Type_of_Building_others.sum(axis=1)

titles.append("Type of a Building: Others")
list_of_tables.append(Type_of_Building_others)
```

```python
Type_of_Building_others
```

Building Category

```python
Building_cat = data['Building Category'].value_counts()

Building_Category = pd.DataFrame(Building_cat)
Building_Category = Building_Category.transpose()

Building_Category['Total'] = Building_Category.sum(axis=1)

titles.append("Building Category")
list_of_tables.append(Building_Category)
```

```python
Building_Category
```

```python
BC_Others = data['Building Category (Other (please specify)) - specify'].value_counts()

Building_Category_others = pd.DataFrame(BC_Others)
Building_Category_others = Building_Category_others.transpose()

Building_Category_others['Total'] = Building_Category_others.sum(axis=1)

titles.append("Building Category: Others")
list_of_tables.append(Building_Category_others)
```

```python
Building_Category_others
```

Type of Buildings VS Building Category

```python
TOB_vs_BC  = data.pivot_table(index='What is the type of building?', columns=['Building Category'], aggfunc='size', fill_value=0)
Type_of_Building_vs_Building_Category = pd.DataFrame(TOB_vs_BC)

# Adding Total

Type_of_Building_vs_Building_Category['Total'] = Type_of_Building_vs_Building_Category.sum(axis=1)
Type_of_Building_vs_Building_Category.loc['Total'] = Type_of_Building_vs_Building_Category.sum(axis=0)

# Adding to final dataframe to be downloaded

titles.append("Type of Building vs Building Category")
list_of_tables.append(Type_of_Building_vs_Building_Category)
```

```python
Type_of_Building_vs_Building_Category
```

Building Category vs Use

```python
BC_vs_use = data.pivot_table(index=['Building Category','Uses of Holding'],aggfunc='size', fill_value=0)
Use_of_Building_vs_Building_Category = pd.DataFrame(BC_vs_use)

# Adding to final dataframe to be downloaded

titles.append("Use of Building vs Building Category")
list_of_tables.append(Use_of_Building_vs_Building_Category)
```

```python
Use_of_Building_vs_Building_Category
```

Use of Building vs Number of Floors

```python
NOF_vs_use = data.pivot_table(index='No of floors/storey of the building', columns=['Uses of Holding'], aggfunc='size', fill_value=0)
Use_of_Building_vs_No_of_floors = pd.DataFrame(NOF_vs_use)

# Adding Total

Use_of_Building_vs_No_of_floors['Total'] = Use_of_Building_vs_No_of_floors.sum(axis=1)

# Adding to final dataframe to be downloaded

titles.append("Use of Building vs Number of Floors")
list_of_tables.append(Use_of_Building_vs_No_of_floors)
```

```python

Use_of_Building_vs_No_of_floors
```

Building Category vs Use of Holding vs No of Floors

```python
BC_No_of_floors_use = data.pivot_table(index=['Building Category', 'Uses of Holding'], columns=['No of floors/storey of the building'], aggfunc='size')
No_of_floors_vs_Building_Category_vs_Use_of_Building_all = pd.DataFrame(BC_No_of_floors_use)

# Adding Total

No_of_floors_vs_Building_Category_vs_Use_of_Building_all['Total'] = No_of_floors_vs_Building_Category_vs_Use_of_Building_all.sum(axis=1)

# Adding to final dataframe to be downloaded

titles.append("Number of Floors vs Building Category vs Use of Holding")
list_of_tables.append(No_of_floors_vs_Building_Category_vs_Use_of_Building_all)
No_of_floors_vs_Building_Category_vs_Use_of_Building_all = No_of_floors_vs_Building_Category_vs_Use_of_Building_all.fillna("")
```

```python
No_of_floors_vs_Building_Category_vs_Use_of_Building_all
```

Residential vs No of floors

```python
desired_category = ['Residential ']
df_residential = data[data['Building Category'].isin(desired_category)]
```

```python
df_residential.shape
```

```python
df_residential.columns
```

```python
red_No_of_floors = df_residential.pivot_table(index='Building Category', columns=['No of floors/storey of the building'], aggfunc='size', fill_value=0)
No_of_floors_vs_Building_Category = pd.DataFrame(red_No_of_floors)

# Adding Total

No_of_floors_vs_Building_Category['Total'] = No_of_floors_vs_Building_Category.sum(axis=1)

# Adding to final dataframe to be downloaded

titles.append("Number of Floors vs Building Category")
list_of_tables.append(No_of_floors_vs_Building_Category)
```

```python
No_of_floors_vs_Building_Category
```

```python
df_other  = df_residential['No of floors/storey of the building (If number of floor > 12, mention the number) - specify'].value_counts()

No_of_floors_others = pd.DataFrame(df_other)

No_of_floors_others.loc['Total'] = No_of_floors_others.sum(axis=0)
No_of_floors_others = No_of_floors_others.transpose()

titles.append("No of floors: Others")
list_of_tables.append(No_of_floors_others)
```

```python
No_of_floors_others
```

Residential vs use vs floors

```python
red_No_of_floors_use = df_residential.pivot_table(index=['Building Category', 'Uses of Holding'], columns=['No of floors/storey of the building'], aggfunc='size', fill_value=0)
No_of_floors_vs_Building_Category_vs_Use_of_Building = pd.DataFrame(red_No_of_floors_use)

# Adding Total

No_of_floors_vs_Building_Category_vs_Use_of_Building['Total'] = No_of_floors_vs_Building_Category_vs_Use_of_Building.sum(axis=1)

# Adding to final dataframe to be downloaded

titles.append("Number of Floors vs Building Category vs Use of Holding")
list_of_tables.append(No_of_floors_vs_Building_Category_vs_Use_of_Building)
```

```python
import pandas as pd

# Create a Pandas Excel writer
with pd.ExcelWriter('BIS_Analysis_Sheet(W27).xlsx', engine='openpyxl') as writer:

    # Initialize a variable to keep track of the current row
    current_row = 1

    # Write each dataframe with its heading
    for i, (df, heading) in enumerate(zip(list_of_tables , titles), start=1):
        # Create or select the worksheet
        sheet_name = 'Sheet1'
        worksheet = writer.sheets.get(sheet_name, None)

        if worksheet is None:
            worksheet = writer.book.create_sheet(sheet_name)

        # Write the heading for the dataframe
        worksheet.cell(row=current_row, column=1, value=heading)

        # Write the dataframe to the worksheet
        start_row = current_row + 1
        df.to_excel(writer, sheet_name=sheet_name, startrow=start_row, index=True)

        # Update the current row for the next dataframe
        current_row = worksheet.max_row + 3

# The Excel file 'output.xlsx' now contains all dataframes with custom headings in the same sheet

from google.colab import files
files.download("BIS_Analysis_Sheet(W27).xlsx")
```
